```python
from openai import OpenAI
import base64
```

```python
client = OpenAI(api_key="123", base_url="https://violet-falcons-yawn.loca.lt/v1")
```

```python
model = "reducto/RolmOCR-7b"
```

```python
def encode_image(image_path):
    with open(image_path, "rb") as image_file:# read only binary format
        return base64.b64encode(image_file.read()).decode("utf-8") #openAI ve benzeri API ler base64 kullanır
```

```python
def ocr_page_with_rolm(img_base64):
    response = client.chat.completions.create( model=model, temperature=0.2, #bu değer modelin daha kesin yanıtlar vermesini sağlar
        max_tokens=4096,
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "image_url",
                        "image_url": {"url": f"data:image/png;base64,{img_base64}"},
                    },
                    {
                        "type": "text",
                        "text": "Return the plain text representation of this document as if you were reading it naturally.\n",
                    },
                ],
            }
        ]
    )
    return response.choices[0].message.content
```

```python
test_img_path = "/content/image.jpg"
img_base64 = encode_image(test_img_path)
print(ocr_page_with_rolm(img_base64))
```
